### The three extensions every data scientist installs first

1. **Python** (Microsoft), enables IntelliSense, linting, debugging, and `conda`/`venv` environment selection
2. **Jupyter** (Microsoft), runs `.ipynb` notebooks natively inside VS Code; no need to open a separate browser
3. **Pylance** (Microsoft), adds fast type-checking and rich autocomplete based on type hints

## Real-world example: Developer productivity and keyboard shortcuts

The chart below shows estimated time savings from using VS Code keyboard shortcuts versus mouse-driven navigation, based on a study of developer workflows. The savings compound across hundreds of repetitions per day.

Notice:

- **Notice:** The largest savings come from operations performed most frequently: opening files, navigating within them, and running code, not from exotic shortcuts
- **Notice:** Time savings are measured in seconds per operation, which seems small until you multiply by 50-200 repetitions per hour during active coding
- **Notice:** The "vs mouse" baseline is not just the travel time to the mouse; it includes visual search, clicking, and recovering your train of thought, which makes the gap larger than it first appears

> **Discussion question:** A classmate says they prefer using the mouse because "shortcuts are hard to remember." What would you tell them about how long it actually takes to internalize the five most valuable shortcuts, and what strategy would you suggest for building the habit?

````python
# ── Keyboard shortcut time savings vs mouse navigation ────────────────────────
tasks = [
    "Open a file by name",
    "Run all cells (notebook)",
    "Comment/uncomment a line",
    "Go to a symbol/function",
    "Open the terminal",
    "Find and replace",
    "Duplicate a line",
    "Move a line up/down",
]
# Estimated seconds saved per operation (shortcut vs. mouse + menu)
time_saved = [4.2, 3.8, 2.1, 5.5, 2.8, 3.3, 1.8, 2.2]
sorted_pairs = sorted(zip(time_saved, tasks), reverse=True)
ts_s, task_s = zip(*sorted_pairs)

fig = go.Figure(go.Bar(
    x=list(ts_s), y=list(task_s),
    orientation="h",
    marker_color=PALETTE["primary"],
    text=[f"{t:.1f}s" for t in ts_s], textposition="outside",
))
layout = base_layout(
    title="Estimated Time Saved Per Operation — Keyboard Shortcut vs Mouse",
    xaxis_title="Seconds Saved Per Use",
    yaxis_title="",
)
layout.update(yaxis=dict(autorange="reversed"), xaxis=dict(range=[0, 7]))
fig.update_layout(layout)
fig.show()
````

### Most valuable VS Code shortcuts for data science work

| Shortcut (Mac) | Shortcut (Win/Linux) | What it does | When to use it |
|---------------|---------------------|-------------|---------------|
| Cmd+P | Ctrl+P | Open any file by name (fuzzy search) | Fastest way to jump between notebooks and scripts |
| Cmd+Shift+P | Ctrl+Shift+P | Open Command Palette | Run any VS Code action by typing its name |
| Ctrl+\` | Ctrl+\` | Toggle integrated terminal | Switch between code and terminal without leaving VS Code |
| Cmd+/ | Ctrl+/ | Comment or uncomment selected lines | Temporarily disable code during debugging |
| Shift+Enter | Shift+Enter | Run current cell (Jupyter) | Execute a cell and move to the next one |
| Cmd+D | Ctrl+D | Select next occurrence of current word | Rename a variable across multiple uses |
| Option+↑/↓ | Alt+↑/↓ | Move line up or down | Reorder imports, reorder function arguments |
| Ctrl+G | Ctrl+G | Go to line number | Jump to the line in a traceback |
| Cmd+Shift+F | Ctrl+Shift+F | Search across all files | Find where a function or variable is used |
| F2 | F2 | Rename symbol (Python-aware) | Rename a variable or function everywhere it is used |
| Cmd+B | Ctrl+B | Toggle side panel | More screen space for code when you need to focus |
| Cmd+1/2/3 | Ctrl+1/2/3 | Focus editor group 1/2/3 | Switch between split editor panes |
