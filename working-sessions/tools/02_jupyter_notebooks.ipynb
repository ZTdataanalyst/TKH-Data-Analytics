## What's happening?

A Jupyter notebook consists of **cells** and a **kernel**. The kernel is a running Python process that holds all variables in memory. Cells are units of input (code or markdown) that communicate with the kernel.

| Concept | What it means | Why it matters |
|---------|-------------|---------------|
| Edit mode | Cursor is inside a cell; typing enters text | Press Enter to enter; Esc to exit |
| Command mode | No cursor in a cell; keypresses run actions | Press Esc to enter; arrow keys move between cells |
| Cell execution order | Cells run in the order *you run them*, not top-to-bottom automatically | Out-of-order execution is the most common source of notebook bugs |
| Kernel state | All variables live in the kernel until restart | A variable from cell 10 is visible in cell 2 if you ran cell 10 first |
| `[*]` indicator | The kernel is currently executing this cell | A long-running cell will show `[*]` until it finishes |

### The most important rule in Jupyter

**Restart the kernel and run all cells before sharing any notebook.** This is the only way to guarantee your notebook runs correctly top to bottom, without relying on hidden state from earlier interactive sessions. Notebooks that work interactively but fail on a clean run are extremely common and completely avoidable.

## Real-world example: How data scientists use notebooks

Data scientists use Jupyter notebooks as their primary workspace for exploration, and then often refactor important code into `.py` modules. The chart below shows how notebook usage breaks down across typical data science work.

Notice:

- **Notice:** Exploratory analysis is the primary use case, where the interactive cell-by-cell model is genuinely superior to running a plain Python script
- **Notice:** Stakeholder communication is a significant share; notebooks export cleanly to HTML and PDF, making them a presentation format as well as a coding environment
- **Notice:** Production code is a small share because notebooks have limitations in production: they are hard to version control cleanly, hard to import from other modules, and hard to test automatically

### The 10 most useful Jupyter keyboard shortcuts

| Feature | Shortcut (Mac) | Shortcut (Windows/Linux) | What it does |
|---------|---------------|--------------------------|-------------|
| Run cell, stay | Ctrl+Enter | Ctrl+Enter | Execute the current cell; cursor stays |
| Run cell, advance | Shift+Enter | Shift+Enter | Execute cell and move to the next one |
| Run cell, insert below | Alt+Enter | Alt+Enter | Execute cell and insert a new one below |
| Enter edit mode | Enter | Enter | Start editing the selected cell |
| Enter command mode | Esc | Esc | Leave the cell editor; enable command shortcuts |
| Insert cell above | A (command mode) | A (command mode) | Add a new cell above current |
| Insert cell below | B (command mode) | B (command mode) | Add a new cell below current |
| Delete cell | D, D (command mode) | D, D (command mode) | Double-tap D to delete selected cell |
| Toggle cell type | M (markdown), Y (code) | M / Y (command mode) | Switch between markdown and code |
| Show shortcuts | H (command mode) | H (command mode) | Open the full keyboard shortcut reference |
