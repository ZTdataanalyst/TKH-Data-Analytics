## What's happening?

The command line (shell) is a text interface to your operating system. Instead of clicking icons to move files, you type short commands. The shell reads your input, asks the OS to do something, and prints the result.

| Command | What it does | Example |
|---------|-------------|---------|
| `pwd` | Print working directory: show where you are | `$ pwd` → `/Users/alice/ds_project` |
| `ls` | List files in the current directory | `$ ls -la` → detailed list including hidden files |
| `cd` | Change directory | `$ cd data/raw/` |
| `mkdir` | Make a new directory | `$ mkdir -p data/processed/` |
| `cp` | Copy a file or directory | `$ cp data/raw/sales.csv data/backup/` |
| `mv` | Move or rename a file | `$ mv old_name.csv new_name.csv` |
| `rm` | Remove a file (no recycle bin: be careful) | `$ rm temp_file.csv` |
| `cat` | Print a file's contents to the terminal | `$ cat environment.yml` |
| `head` | Print the first n lines of a file | `$ head -20 data.csv` |
| `grep` | Search for a pattern inside files | `$ grep "error" pipeline.log` |

### Absolute vs relative paths

An **absolute path** starts from the root of the file system: `/Users/alice/ds_project/data/raw/sales.csv`. It works from anywhere.

A **relative path** is relative to your current directory: `data/raw/sales.csv` only works if you are already inside `/Users/alice/ds_project/`. This is why `pd.read_csv("data/raw/sales.csv")` in your notebook depends on where the notebook is located.

Almost every action in the lifecycle of a data science project touches the terminal. The chart shows how often different command-line operations appear in a realistic project from setup to deployment.

Notice:

### The 15 most important commands for data science workflows

| Command | What it does | Data science use case |
|---------|-------------|----------------------|
| `pwd` | Print current directory | Confirm where you are before running scripts |
| `ls -la` | List all files with details | Check file sizes before downloading/moving large datasets |
| `cd path/` | Change directory | Navigate to project root before running code |
| `mkdir -p dir/` | Create directory (and parents) | Create `data/raw/`, `data/processed/` at project start |
| `cp src dst` | Copy file | Back up a dataset before modifying it |
| `mv src dst` | Move or rename | Rename a file without creating a copy |
| `rm file` | Delete file permanently | Clean up temporary files (no undo!) |
| `cat file` | Print file contents | Quickly inspect a small CSV or config file |
| `head -n 20 file` | First 20 lines | Check the header and first rows of a large CSV |
| `grep pattern file` | Search for text | Find error messages in a log file |
| `wc -l file` | Count lines | Check how many rows a CSV has |
| `python script.py` | Run a Python script | Execute a training or preprocessing script |
| `conda activate env` | Activate an environment | Switch to the correct Python for a project |
| `pip install pkg` | Install a Python package | Add a library to the active environment |
| `which python` | Show path to active Python | Verify the right Python is active |
