## What's happening?

A programming language is a set of rules for writing instructions that a computer can eventually execute. The key distinction is *when* those instructions are translated into something the machine can run.

| Execution model | How it works | Example languages |
|----------------|-------------|------------------|
| Compiled | Source code is translated to machine code before running; the resulting binary runs directly on the CPU | C, C++, Rust, Go |
| Interpreted | Source code is read and executed line by line at runtime by an interpreter program | Python, R, Ruby |
| Bytecode compiled + VM | Source is compiled to intermediate bytecode at runtime, then run by a virtual machine | Java, Kotlin, Python (CPython also does this internally) |

Python specifically uses **CPython**, the reference implementation. When you run `python script.py`, CPython reads your source, compiles it to bytecode (stored in `.pyc` files), then executes that bytecode on the Python Virtual Machine. The key point is that all of this happens automatically when you press Run, with no separate compile step you need to manage.

### What "interpreted" means in practice

Because Python does not require a compilation step, you get instant feedback: write a line, run it, see the result. This is exactly why Jupyter notebooks work the way they do. The downside is speed: Python code typically runs 10-100x slower than equivalent C++ code, which is why performance-critical operations in NumPy and pandas are actually written in C under the hood and just exposed through Python.

## Real-world example: Why Python dominates data science

The chart below shows language usage among data scientists and ML practitioners, drawn from the 2023 Stack Overflow Developer Survey and the Kaggle State of Data Science Report. This is not just a popularity contest: the numbers reflect which tools the ecosystem has built up around, which languages employers hire for, and which ones have the best libraries for the work.

Notice:

- **Notice:** Python's lead is not marginal; it is used by the vast majority of data science practitioners, which means the tutorials, the libraries, and the job postings all assume Python
- **Notice:** R is a strong second specifically in statistics and academic research, not in engineering or production ML; many data scientists know both
- **Notice:** SQL appears high on the list not because it is a general-purpose programming language but because data almost always lives in a database before it reaches Python

> **Discussion question:** A colleague argues that Julia is faster than Python and equally readable, so data scientists should switch. What would it actually cost to make that switch for a team of 10 who have two years of Python code, trained models, and production pipelines?

```python
# ── Language usage among data science practitioners (2023 survey data) ───────────
languages = ["Python", "SQL", "R", "JavaScript", "Bash/Shell",
             "Java", "C/C++", "Julia", "Scala", "MATLAB"]
pct_ds = [94.3, 71.2, 38.1, 21.4, 33.6, 17.2, 12.8, 5.4, 9.1, 7.3]
colors = [
    PALETTE["primary"] if lang == "Python" else
    PALETTE["accent"]  if lang == "R"      else
    PALETTE["muted"]
    for lang in languages
]
sorted_pairs = sorted(zip(pct_ds, languages, colors), reverse=True)
pct_s, lang_s, col_s = zip(*sorted_pairs)

fig = go.Figure(go.Bar(
    x=pct_s, y=lang_s,
    orientation="h",
    marker_color=list(col_s),
    text=[f"{p:.0f}%" for p in pct_s], textposition="outside",
))
layout = base_layout(
    title="Language Usage Among Data Scientists & ML Practitioners (2023)",
    xaxis_title="% of Respondents Who Use Regularly",
    yaxis_title="",
)
layout.update(yaxis=dict(autorange="reversed"), xaxis=dict(range=[0, 105]))
fig.update_layout(layout)
fig.show()
```

### Language comparison for data science work

| Language | Compiled or Interpreted | Common data science use | Why data science uses or avoids it |
|----------|------------------------|------------------------|------------------------------------|
| Python | Interpreted (CPython bytecode VM) | Everything: EDA, modeling, deployment | Vast ecosystem, readable, interactive |
| R | Interpreted | Statistical analysis, academic research | Best-in-class stats packages; weaker for production |
| SQL | Interpreted (query engine) | Data extraction, aggregation in databases | Essential complement to Python, not a replacement |
| Julia | JIT compiled | High-performance numerical computing | Fast, but small ecosystem and community |
| Java / Scala | Bytecode compiled + JVM | Big data (Spark) | Verbose; used when Python is too slow at scale |
| C / C++ | Compiled to machine code | Low-level ML inference, embedded systems | Fast, but complex; used inside libraries not by end users |

## Key takeaway

> **Python is interpreted, readable, and slow at raw computation but fast at data science work because the heavy lifting happens inside C-written libraries like NumPy; that combination is why it became the language of the field.**
